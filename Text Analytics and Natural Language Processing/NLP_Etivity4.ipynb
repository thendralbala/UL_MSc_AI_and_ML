**Task 1 (B)**

Write a simple spelling correction function that takes a nonWord string as input (i.e. non-word misspelling such as “acress”), and prints out a list of candidate corrections.
Hint 1: only use the language model/priors, i.e. the unigram counts stored in unigrams_df, and Ignore the channel/error model probabilities.
Hint 2: use the Damerau-Levenshtein Distance function in the jellyfish library for candidate generation.


```python
def nonWordSpellingCorrection(nonWord):
  candidates = []
  print(f"Candidate Corrections for {nonWord}:")
  print(" Word      Count          Probability")
  total_count = unigrams_df['count'].sum()  # Calculate total word count for probability
  for word in unigrams_df['unigram']:
    distance = jellyfish.damerau_levenshtein_distance(nonWord, str(word))
    if distance <= 2:
      word_count = unigrams_df.loc[unigrams_df['unigram'] == word]['count'].values[0]
      probability = word_count / total_count  # Calculate word probability
      print(word, word_count, probability)
      candidates.append(
          (word, word_count, distance, probability)  # Append probability to tuple
      )
  return candidates

nonWord = "acress"
candidate_corrections = nonWordSpellingCorrection(nonWord)
```

Use the TextBlob Library (https://textblob.readthedocs.io/en/dev/) to write a function that takes a string (e.g.  acress) as input and returns a list of (word, confidence) tuples with spelling suggestions.


```python
from textblob import Word

def func(word):

  w = Word(word)
  return w.spellcheck()


print(func("acress"))
```

Use the TextBlob Library (https://textblob.readthedocs.io/en/dev/) to write a function that takes a sentence as input and corrects its spelling.

```python
from textblob import TextBlob

def correct_spelling(sentence):
    blob = TextBlob(sentence)
    corrected_sentence = str(blob.correct())
    return corrected_sentence

sentence = "I havv good speling!"
corrected_sentence = correct_spelling(sentence)
print(f"Sentence = {sentence}")
print(f"Corrected Sentence = {corrected_sentence}")
```
