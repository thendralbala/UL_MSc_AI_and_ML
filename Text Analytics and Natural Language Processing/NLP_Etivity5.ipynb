#Task 1: Sentiment Analysis
Sentiment analysis is the interpretation and classification of emotions within text. Write a simple rule-based sentiment analyzer, which takes a string as input and returns its sentiment (positive, negative, neutral).
Here’s a basic example of how a rule-based system works:

Define two lists of polarized words (e.g. negative words such as bad, worst, ugly, and positive words such as good, best, beautiful).

HINT: you can use Bing Liu's lists of positive and negative sentiment words (RAR archive). You should directly wget them to your notebook from: http://ptrckprry.com/course/ssd/data/positive-words.txt and http://ptrckprry.com/course/ssd/data/negative-words.txt

Count the number of positive and negative words that appear in a given text.
If the number of positive word appearances is greater than the number of negative word appearances, the system returns a positive sentiment and vice versa. If the numbers are even, the system will return a neutral sentiment.


```python
import pandas as pd


!wget http://ptrckprry.com/course/ssd/data/positive-words.txt #Opinion Lexicon: Positive Words; source: http://www.cs.uic.edu/~liub/FBS/sentiment-analysis.html
!wget http://ptrckprry.com/course/ssd/data/negative-words.txt #Opinion Lexicon: Negative Words; soruce: http://www.cs.uic.edu/~liub/FBS/sentiment-analysis.html


filePath1 = "/content/positive-words.txt"
filePath2 = "/content/negative-words.txt"


posWords_df = pd.read_csv(filePath1,sep='\t',header=None, names=['words'],encoding='latin-1',skiprows=[i for i in range(0,34)]) #Latin-1 (ISO-8859-1) character set. use chardet lib if needed
negWords_df = pd.read_csv(filePath2,sep='\t',header=None, names=['words'],encoding='latin-1',skiprows=[i for i in range(0,34)])


print(f'Number of posWords: {len(posWords_df)} Number of negWords: {negWords_df.size}')
display(posWords_df.head(10))
display(negWords_df.head(10))
```

```python
def sentimentAnalyzer(text):
  posCount = 0
  negCount = 0
  for word in text.split():
    if word in posWords_df['words'].values:
      posCount += 1
    elif word in negWords_df['words'].values:
      negCount += 1

  if posCount > negCount:
    result =  'Positive Sentiment :)'
  elif negCount > posCount:
    result = 'Negative Sentiment :('
  else:
    result = 'Neutral Sentiment :|'

  confidence = (posCount + negCount) / len(text.split())

  print(f"Sentence = {text}")
  print(result)
  print(f"Confidence: {confidence}")
  print("--------------------------------------------------------------------")




sentimentAnalyzer('NLP is cool')
sentimentAnalyzer('NLP is cool and useful')
sentimentAnalyzer('NLP is hard')
sentimentAnalyzer('NLP is hard and useless')
sentimentAnalyzer('NLP stands for Natural Language Processing')
```

#Task 2: Naive Bayes Classifier

Write a NaiveBayes function which takes the sample training and testing documents shown in the table below as input, and uses the Naive Bayes algorithm to classify the test documents.

Hint1: Python Dictionaries are suitable data structures for storing BOWs (Bag Of Words)


Sample output:








```python
def calculate_word_conditional_prob(word, megaDoc, classCount):
    wordCount = megaDoc.lower().split().count(word.lower())
    vocabSize = len(set(megaDoc.lower().split()))
    return (wordCount + 1) / (len(megaDoc.lower().split()) + vocabSize) #Laplace smoothing
```

```python
def naiveBayesClassifier(trainingSet, testSet):

  #megaDocGB contains the sentences from training set whose classes are GB
  megaDocGB = ''
  #megaDocIE contains the sentences from training set whose classes are IE
  megaDocIE = ''

  # Iterate through the training set and populate megaDocGB and megaDocIE
  for sentence, classLabel in trainingSet:
    if classLabel == 'GB':
      megaDocGB += sentence + ' '  # Add the sentence to megaDocGB
    elif classLabel == 'IE':
      megaDocIE += sentence + ' '  # Add the sentence to megaDocIE

  print(f'megaDocGB: {megaDocGB}')
  print(f'megaDocIE: {megaDocIE}')


  probGB = len([item for item in trainingSet if item[1] == 'GB']) / len(trainingSet)
  probIE = len([item for item in trainingSet if item[1] == 'IE']) / len(trainingSet)

  print(f'probGB: {probGB}')
  print(f'probIE: {probIE}')

  #GB Bag of Words
  GB_bow = {}
  #IE Bag of Words
  IE_bow = {}

  for sentence, classLabel in trainingSet:
    for word in sentence.split():
      if classLabel == 'GB':
        if word in GB_bow:
          GB_bow[word] += 1
        else:
          GB_bow[word] = 1

      elif classLabel == 'IE':
        if word in IE_bow:
          IE_bow[word] += 1
        else:
          IE_bow[word] = 1

  print(f'GB_bow: {GB_bow}')
  print(f'IE_bow: {IE_bow}')

  #The volacabulary
  V = set(megaDocGB.split() + megaDocIE.split())
  print(f'V: {V}')
  print(f'|V| : {len(V)}')

  for testDocument, _ in testSet:
    print("--------------------------------------------------------------------------------------------------------------------------------------")
    print(f"Test Document: {testDocument}\n") # Print the test document
    docProbabilityGB = probGB  # Initialize with prior probability for GB
    docProbabilityIE = probIE  # Initialize with prior probability for IE

    # Iterate through words in the test document and print probabilities
    for word in testDocument.split():
      wordConditionalProbGB = calculate_word_conditional_prob(word, megaDocGB, probGB)
      wordConditionalProbIE = calculate_word_conditional_prob(word, megaDocIE, probIE)

      print(f"  Word: {word}       Conditional Probability (GB): {wordConditionalProbGB}       Conditional Probability (IE): {wordConditionalProbIE}")
      docProbabilityGB *= calculate_word_conditional_prob(word, megaDocGB, probGB)
      docProbabilityIE *= calculate_word_conditional_prob(word, megaDocIE, probIE)


    print(f"\n  Document Probability (GB): {docProbabilityGB}       Document Probability (IE): {docProbabilityIE}")
    inferredClass = 'GB' if docProbabilityGB > docProbabilityIE else 'IE'

    print(f"  Inferred Class: {inferredClass}")
```

```python
trainingSet = [('London is the Capital of GB','GB'),('Oxford is a city in GB','GB'),('Dublin is the capital of Ireland','IE'),('Limerick is a city in Ireland','IE')]
testSet = [('University of Limerick','?'),('University College Dublin','?'),('Imperial College London','?'),('University of Oxford','?'),('Ireland & GB','?')]
naiveBayesClassifier(trainingSet,testSet)
```
