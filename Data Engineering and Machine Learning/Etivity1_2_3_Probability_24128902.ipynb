## Problem 1: Lottery Draw

You are participating in a lottery where a ball is drawn from a set of balls numbered from 1 to 10. Each number has an equal chance of being drawn. The lottery pays out different amounts based on the number drawn:


*  Number 1-3: Prize of €2
*  Number 4-6: Prize of €5
*   Number 7-9: Prize of €10
*   Number 10: Prize of €50

Simulate 1000 lottery draws and:


1.   Calculate the expected value and standard deviation of the lottery prize.
2.   Plot the frequency distribution of the lottery prizes.







**Approach**

To solve Problem 1, the following approach was taken:

1. **Define the Prize Distribution**:
   - A set of lottery prizes in euros was defined based on the ball number drawn:
     - Numbers 1-3: €2
     - Numbers 4-6: €5
     - Numbers 7-9: €10
     - Number 10: €50
   - This was implemented by creating a list of prize values where each value corresponds to a specific ball number.

2. **Simulate Lottery Draws**:
   - To simulate 1000 lottery draws, `numpy.random.randint` was used. This function randomly selects from the range of ball numbers (1-10, both includive)

3. **Calculate Statistics**:
   - The expected value (mean) and standard deviation of the lottery prizes were computed using `numpy.mean` and `numpy.std`, respectively. This quantifies the average prize and the spread of prize amounts around this average.

4. **Visualize the Results**:
   - A histogram was plotted using `matplotlib.pyplot.hist` to visualize the frequency distribution of the prizes. The histogram displayed how often each prize amount occurred out of the 1000 draws. The bins for the histogram were chosen to represent the prize categories clearly.



```python
import numpy as np
import matplotlib.pyplot as plt

# Number of simulations
n_simulations = 1000

# Simulating lottery draws (1-10)
lottery_draws = np.random.randint(1, 11, n_simulations)

# Calculating lottery prizes based on the draw
# Define prize mapping
prize_mapping = {
    1: 2, 2: 2, 3: 2,
    4: 5, 5: 5, 6: 5,
    7: 10, 8: 10, 9: 10,
    10: 50
}

# Assign prizes based on the draw
lottery_prizes = np.array([prize_mapping[draw] for draw in lottery_draws])

# Step 3: Calculate expected values and standard deviations
expected_prize = np.mean(lottery_prizes)
std_dev_prize = np.std(lottery_prizes)

# Displaying results
print(f"Expected value of lottery prizes: €{expected_prize:.2f}")
print(f"Standard deviation of lottery prizes: €{std_dev_prize:.2f}")

# Step 4: Plotting the frequency distribution
plt.figure(figsize=(8, 5))
plt.hist(lottery_prizes, bins=[1.5, 2.5,4.5, 5.5, 9.5,10.5,49.5,50.5], density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Frequency Distribution of Lottery Prizes')
plt.xlabel('Lottery Prize (€)')
plt.ylabel('Frequency')
plt.xticks([2, 5, 10, 50])

plt.grid(axis='y')
plt.show()
```

##Problem 2: Reward Based Fitness Challenge
Consider a fitness challenge where participants' performance, measured in miles run during a month, is normally distributed with a mean of 30 miles and a standard deviation of 5 miles. The organising committee plans to reward participants based on their performance:


*   No reward for running less than 25 miles
*   $10 reward for running between 25 and 35 miles

*   $25 reward for running between 35 and 50 miles


Simulate the total rewards distributed to 5,000 participants, calculate the probabilities for each reward category based on the normal distribution, and assess the average cost incurred per participant in order to analyze the effectiveness of this incentive structure.






**Approach**



1.   **Define Parameters for the Normal Distribution:**

      Define the normal distribution parameters: mean, standard deviation and the total number of participants (the total number of simulations to be made)
2.   **Simulate Running Distances:**

      Set the seed value to 0 for reproducibility. Use the numpy.random.normal function to simulate the running distances of all the participants based on the normal distribution parameters.

3.   **Define Reward Categories**

      Use the elif ladder method to define a function that assigns a reward value to each running distance which is taken as the input for the function. Store the reward values for all the simulated running distances in a list.
      The sum of all elements of the rewards list is the total amount spent in the rewarding the participants. This sum when divided by the total number of participants gives the average reward per person.
4.   **Visualize the Results:**

      The visualization details involve creating a histogram of the simulated running distances using Seaborn, with 30 bins for a detailed distribution view. A kernel density estimate (KDE) is overlaid to provide a smooth representation of the data. Vertical dashed lines indicate the reward thresholds at 25, 35, and 50 miles, with distinct colors for clarity.







```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters for the normal distribution
mean = 30
std_dev = 5
num_participants = 5000

# Simulate running distances
np.random.seed(0)  # For reproducibility
distances = np.random.normal(mean, std_dev, num_participants)

# Define reward categories
def calculate_reward(distance):
    if distance < 25:
        return 0
    elif 25 <= distance < 35:
        return 10
    elif 35 <= distance <= 50:
        return 25
    else:
        return 0  # No reward for distances > 50

# Calculate rewards for all participants
rewards = [calculate_reward(d) for d in distances]

# Total and average rewards
total_reward = sum(rewards)
average_reward_per_participant = total_reward / num_participants

print(f"Total Reward Distributed: ${total_reward}")
print(f"Average Reward per Participant: ${average_reward_per_participant:.2f}")

# Plotting the distribution of distances
plt.figure(figsize=(10, 6))
sns.histplot(distances, bins=30, kde=True, color='blue', stat='density')
plt.title('Distribution of Running Distances in the Fitness Challenge')
plt.xlabel('Miles Run')
plt.ylabel('Density')
plt.axvline(25, color='red', linestyle='--', label='25 miles threshold')
plt.axvline(35, color='green', linestyle='--', label='35 miles threshold')
plt.axvline(50, color='orange', linestyle='--', label='50 miles threshold')
plt.legend()
plt.grid()
plt.show()
```

```python
X = data.iloc[:,1:3] # X is the features in our dataset
y = data.iloc[:,0]   # y is the Labels in our dataset
```

### Train/Test Split

We have a bunch of data from the csv file. What we need to do is decide what of that data we will use to train the algorithm, and what of that data will be used to see how good the result is. More on this later in the module, but you can't use the same data you train with to test the outcome, as it will naturally fit that very well, but you won't catch things like overfitting.

There is some science to this, but for now we will randomly select one third for testing, and the rest for training.

```python
################################
from sklearn.model_selection import train_test_split

#Define features and Target
X = data[['Height','Weight']]
y = data['Gender']

# Split the data into training and test sets (33% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)

```

### Train the algorithm
We won't go into the details here, but we will use a Gaussian Naive Bayesian. Use GaussianNB from sklearn to train the model. The Gaussian is needed as we're dealing with sparse, continuous data

```python
from sklearn.naive_bayes import GaussianNB

################################
model = GaussianNB()
model.fit(X_train, y_train)
```

### Test the algorithm
Using the test data

```python
from sklearn.metrics import accuracy_score

################################
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
